```
alfa = 0.01

# Hidden Layer - Neurona 1
w_n1_l1 = np.random.normal(0,1,(3,1)) # dim = 3x1
print('Inicializacion en el leyer 1 de W en la neurona 1: ')
print(w_n1_l1)
print()

x = np.array([[1,0,1]]) # dim 1x3
z1_l1 = x.T * w_n1_l1 # dim 3x1
print('Z1 en el leyer 1 da: ')
print(z1_l1)
print()

a1 = 1/ (1 + np.exp(-1* z1_l1.sum())) # dim 1x1
print('a1 en el leyer 1 da: ')
print(a1)
print()


# Hidden Layer - Neurona 2
w_n2_l1 = np.random.normal(0,1,(3,1))
print('Inicializacion en el leyer 1 de W en la neurona 2: ')
print(w_n2_l1)
print()

x = np.array([[1,0,1]])
z2_l1 = x.T * w_n2_l1
print('Z2 en el leyer 1 da: ')
print(z2_l1)
print()

a2 = 1/ (1 + np.exp(-1* z2_l1.sum()))
print('a2 en el leyer 1 da: ')
print(a2)
print()
# Output Layer - Neurona salida
w_output = np.random.normal(0,1,(3,1))
y_hat = np.array([[a1,a2,1]]) @ w_output
print('y_hat en el output leyer da: ')
print(y_hat)
print()
# Calculo del error 
error = 0 - y_hat

# Actualizacion de los parametros Output Layer

w_output[0] = w_output[0] + alfa * 2 * error * a1
w_output[1] = w_output[1] + alfa * 2 * error * a2
w_output[0] = w_output[2] + alfa * 2 * error 

#w_output = np.array([[w_output[0] + alfa * 2 * error * a1], 
                     #[w_output[1] + alfa * 2 * error * a2], 
                     #[w_output[2] + alfa * 2 * error]])

print('Actualizacion del W en el output layer que da: ')
print(w_output)
print()
# Actualizacion de los parametros del Hidden Layer
w_n1_l1[0] = w_n1_l1[0] + alfa * (-2 * error * w_output[0]) * (a1 * (1 - a1)) * x[0,0]
w_n1_l1[1] = w_n1_l1[1] + alfa * (-2 * error * w_output[1]) * (a1 * (1 - a1)) * x[0,1]
w_n1_l1[2] = w_n1_l1[2] + alfa * (-2 * error * w_output[2]) * (a1 * (1 - a1)) * x[0,2]
w_n2_l1[0] = w_n2_l1[0] + alfa * (-2 * error * w_output[0]) * (a2 * (1 - a2)) * x[0,0]
w_n2_l1[1] = w_n2_l1[1] + alfa * (-2 * error * w_output[1]) * (a2 * (1 - a2)) * x[0,1]
w_n2_l1[2] = w_n2_l1[2] + alfa * (-2 * error * w_output[2]) * (a2 * (1 - a2)) * x[0,2]

print('Actualizacion del W en el hidden layer que da: ')
print(w_n1_l1)
print()
print(w_n2_l1)

```

```
def Stockastic_gradient_decent_net(x, y, lr=0.01, n_epochs=100):
    
    f = x.shape[0]
    c = x.shape[1]
    
    #w = np.random.randn(c).reshape(c,1)
    w_n1_l1 = np.random.normal(0,1,(c,1)) # dim = 3x1
    w_n2_l1 = np.random.normal(0,1,(c,1))
    w_output = np.random.normal(0,1,(c,1))

    for epoch in range(0,n_epochs):
        #idx = np.random.permutation(f)
        #x = x[idx]
        #y = y[idx]
        
        for n in range(0,f):
            # Hidden Layer - Neurona 1
            z1_l1 = x[n].T * w_n1_l1 # dim 3x1
            a1 = 1/ (1 + np.exp(-1* z1_l1.sum())) # dim 1x1

            # Hidden Layer - Neurona 2
            z2_l1 = x[n].T * w_n2_l1
            a2 = 1/ (1 + np.exp(-1* z2_l1.sum()))

            # Output Layer - Neurona salida
            y_hat = np.array([[a1,a2,1]]) @ w_output

            # Calculo del error 
            print('Epoch # {}, n = {}'.format(epoch, n))
            print(y_hat)
            print(y[n])
            error = y[n] - y_hat
            print(error)

            # Actualizacion de los parametros Output Layer

            w_output[0] = w_output[0] + lr * 2 * error * a1
            w_output[1] = w_output[1] + lr * 2 * error * a2
            w_output[0] = w_output[2] + lr * 2 * error 

            # Actualizacion de los parametros del Hidden Layer
            w_n1_l1[0] = w_n1_l1[0] + lr * (-2 * error * w_output[0]) * (a1 * (1 - a1)) * x[n,0]
            w_n1_l1[1] = w_n1_l1[1] + lr * (-2 * error * w_output[1]) * (a1 * (1 - a1)) * x[n,1]
            w_n1_l1[2] = w_n1_l1[2] + lr * (-2 * error * w_output[2]) * (a1 * (1 - a1)) * x[n,2]
            w_n2_l1[0] = w_n2_l1[0] + lr * (-2 * error * w_output[0]) * (a2 * (1 - a2)) * x[n,0]
            w_n2_l1[1] = w_n2_l1[1] + lr * (-2 * error * w_output[1]) * (a2 * (1 - a2)) * x[n,1]
            w_n2_l1[2] = w_n2_l1[2] + lr * (-2 * error * w_output[2]) * (a2 * (1 - a2)) * x[n,2]


    
```
